```python
from holoviews.operation import contours

x = y = np.arange(-3.0, 3.0, 0.1)
X, Y = np.meshgrid(x, y)


def g(x, y, c):
    return 2 * ((x - y) ** 2 / (x ** 2 + y ** 2)) + np.exp(
        -((np.sqrt(x ** 2 + y ** 2) - c) ** 2)
    )


img = hv.Image(g(X, Y, 2))
filled_contours = contours(img, filled=True)

filled_contours
```

```python
holomap = hv.HoloMap(
    [(t, hv.Image(g(X, Y, 4 * np.sin(np.pi * t)))) for t in np.linspace(0, 1, 21)]
).opts(cmap="fire", colorbar=True, show_title=False, xaxis="bare", yaxis="bare")

contour_hmap = contours(holomap, filled=True)

hv.output(contour_hmap, holomap="gif", fps=5)
```
